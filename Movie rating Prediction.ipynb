```python
## remove_breaklines
for i in range(40000):
    X_train[i] = str(X_train[i]).replace('<br /><br />' , '')[2:-2]
        
```

```python
tokenizer = RegexpTokenizer('[a-zA-Z]+')
```

```python
X = []
vocab = set()
for i in range(10000):
    sentence = tokenizer.tokenize(X_train[i][0])
    sentence = [stem.stem(i) for i in sentence if i.lower() not in sw and len(i) > 2]
    for word in sentence:
        vocab.add(word)
    X.append(sentence)
```
